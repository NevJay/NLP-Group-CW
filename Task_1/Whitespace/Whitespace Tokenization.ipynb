```python
import nltk
from nltk.tokenize import WhitespaceTokenizer
```

```python
with open('/content/drive/MyDrive/wikitext-2/wikitext-2/wiki.train.tokens', 'r') as file:
    content1  = file.readlines()

with open('/content/drive/MyDrive/wikitext-2/wikitext-2/wiki.test.tokens', 'r') as file:
    content2  = file.readlines()

with open('/content/drive/MyDrive/wikitext-2/wikitext-2/wiki.valid.tokens', 'r') as file:
    content3  = file.readlines()
```

```python
content = content1 + content2 + content3
```

```python
# Whitespace Tokenization

def whitespace_tokenization(content):
    tokens = content.split()
    return tokens
```

```python
whitespace_tokens = []

for line in content:
    tokens = ' '.join(line).split()
    whitespace_tokens.extend(tokens)

print(whitespace_tokens[0:100])
```

```python
output_file_path = '/content/drive/MyDrive/NLPCoursework/Whitespace/whitespace_data.txt'
```

```python
with open(output_file_path, 'w') as output_file:
    for token in whitespace_tokens:
        output_file.write(token + '\n')
```
