```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
with open('/content/drive/MyDrive/wikitext-2/wikitext-2/wiki.train.tokens', 'r') as file:
    content1  = file.readlines()

with open('/content/drive/MyDrive/wikitext-2/wikitext-2/wiki.test.tokens', 'r') as file:
    content2  = file.readlines()

with open('/content/drive/MyDrive/wikitext-2/wikitext-2/wiki.valid.tokens', 'r') as file:
    content3  = file.readlines()
```

```python
content = content1 + content2 + content3
```

```python
# Lemmatization

def lemmatization(content):
  lemmatizer = WordNetLemmatizer()
  tokens = [lemmatizer.lemmatize(token) for token in content]
  return tokens
```

```python
lemmatized_content = [lemmatization(line.split()) for line in content]
```

```python
output_file_path = '/content/drive/MyDrive/NLPCoursework/Lemmatization/lemmatized_data.txt'
```

```python
with open(output_file_path, 'w') as output_file:
    for tokens in lemmatized_content:
        output_file.write(' '.join(tokens) + '\n')
```

```python
for line in content:
    tokens = lemmatization(line)
    print(tokens[0:100])
```
